# CRAFTs and TEMPLATEs

## Templates

If you have a dialog called 'TEMPLATE' in the current directory, all the messages in that dialog will get added to the start of any new dialogs you create. This can save a lot of time if you find yourself starting with the same preamble/imports/... a lot!

Templates apply to all subfolders too - and they stack! We have a TEMPLATE dialog in this folder, and in the creatively-named `subfolder` folder - any new dialogs created in `subfolder` will start with the contents of both TEMPLATEs.

## CRAFTS

Crafts (written in dialogs called 'CRAFT') are similar, with two key differences: all code in CRAFTS is also executed (so be very careful with them!) and they are added to the AI context but NOT shown in the dialog. Both you and the AI can thus use functions defined in crafts (the AI only if they're also referenced as tools). 



To see a minimal example in action, create a new dialog in this folder or `subfolder` and observe the messages that you start with, then try running `foo()`.
