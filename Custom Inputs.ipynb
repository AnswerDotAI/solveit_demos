## Custom Styling

By default, we detect `input` and render a LabelInput in a form. If you'd like to make something custom, you can return arbitrary HTML wrapped in `<solveit-input>` tags. For it to work, your HTML must:

- Send a POST request to `/input_reply_`
- Specify the `id_` as `__msg_id`
- Specify the `user_input`

For example, here's how you might make a custom Yes/No button widget:

````
ans = input(f"""
<solveit-input>
    <form hx-post="/input_reply_?id_={__msg_id}" hx-swap="none">
        <button type="submit" name="user_input" value="Yes" class='uk-btn uk-btn-primary'>Yes</button>
        <button type="submit" name="user_input" value="No"  class='uk-btn uk-btn-default'>No</button>
    </form>
</solveit-input>
""")
print(ans)
````

Or with FastHTML:

````
from fasthtml.common import *
f = Div(
    Div('Are you sure?', cls='text-sm font-semibold mb-2'),
    Form(
        Button('Yes', type='submit', name='user_input', value='Yes', cls='uk-btn uk-btn-primary'),
        Button('No', type='submit', name='user_input', value='No', cls='uk-btn uk-btn-default'),
        hx_post=f'/input_reply_?id_={__msg_id}',
        hx_swap='none',
        cls='flex gap-2'
    ),
    cls='rounded-xl border border-slate-300 bg-white p-4 shadow-sm'
)
ans = input(f"<solveit-input>{f}</solveit-input>")
print(ans)
````
